## Salary Overview

Salary is one of the primary variables analyzed in this project.

Understanding salary distribution helps identify compensation patterns across AI roles and provides context for later visualizations.

```python
salary_summary = df['annual_salary_usd'].describe()

salary_summary
```

The average annual salary across all AI positions is approximately $195,000, with salaries ranging from $90,000 to $384,000.

The wide salary range suggests substantial variation across AI career paths, indicating that factors such as specialization, experience, and market demand may play an important role in determining compensation.

## Salary by Experience Level

A preliminary analysis was conducted to examine how compensation changes as professionals gain experience within the AI workforce.

```python
experience_salary = (
    df.groupby('experience_level')['annual_salary_usd']
      .mean()
      .sort_values()
)

experience_salary
```

Average salaries increase consistently across all experience levels. Entry-level professionals earn approximately \$150,000 annually, while lead-level professionals earn over $240,000.

This progression highlights the significant financial benefits associated with career advancement in the AI industry and suggests that experience remains a key driver of compensation.

## Highest-Paying AI Roles

To identify the most financially rewarding career paths, average salaries were calculated for each AI job role.

```python
top_roles = (
    df.groupby('job_title')['annual_salary_usd']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

top_roles
```

Several emerging AI specialties appear among the highest-paying positions in the dataset. AI Solutions Architect, Senior ML Engineer, and LLM Engineer all report average salaries exceeding $240,000.

Interestingly, many of the highest-paying roles are associated with advanced AI systems, generative AI, and large language models, reflecting current industry investment and demand for specialized expertise.

## Demand Analysis

Demand scores provide an estimate of market demand for specific AI roles.

Analyzing demand alongside compensation helps identify career paths that offer both strong salaries and favorable employment opportunities.

```python
role_summary = (
    df.groupby('job_title')
      .agg({
          'annual_salary_usd':'mean',
          'demand_score':'mean'
      })
      .reset_index()
)

role_summary.sort_values('demand_score', ascending=False)
```

LLM Engineer, ML Engineer, Senior ML Engineer, and AI Agent Developer demonstrate the highest demand scores within the dataset.

Many of these roles also appear among the highest-paying positions, suggesting that organizations are willing to offer premium compensation to attract talent in rapidly growing areas of artificial intelligence.

## LLM Role Salary Premium

Large Language Model (LLM) roles have emerged as one of the fastest-growing segments of the AI workforce.

A comparison was conducted between LLM-related roles and non-LLM roles to determine whether a measurable salary premium exists.

```python
llm_salary = (
    df.groupby('is_llm_role')['annual_salary_usd']
      .mean()
)

llm_salary
```

LLM-related positions earn approximately \$207,746 annually on average, compared to $191,309 for non-LLM roles.

This difference represents a salary premium of roughly $16,400 per year, indicating that expertise in large language models and generative AI technologies currently carries measurable economic value in the job market.
