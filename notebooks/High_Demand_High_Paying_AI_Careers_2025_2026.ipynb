# Project Overview

Artificial Intelligence continues to reshape industries and create new career opportunities across the global workforce. As organizations increase investments in AI technologies, demand for specialized professionals such as LLM Engineers, AI Agent Developers, Machine Learning Engineers, and AI Architects continues to grow.

This project analyzes the AI Jobs Market 2025–2026 dataset to identify salary trends, demand patterns, and emerging career opportunities within the AI industry. The analysis focuses on understanding which AI roles offer the strongest combination of compensation and market demand.

The findings are intended to help students, professionals, hiring managers, and organizations better understand the evolving AI employment landscape and identify promising career pathways.

# Purpose of the Study

The purpose of this project is to explore the current AI job market and identify trends that influence career opportunities within the field of Artificial Intelligence.

Specifically, this analysis seeks to:

- Examine how salaries change across experience levels.
- Identify the highest-paying AI roles.
- Explore the relationship between salary and market demand.
- Highlight emerging AI positions that offer strong career potential.
- Provide actionable insights for individuals pursuing careers in AI.

# Research Question
