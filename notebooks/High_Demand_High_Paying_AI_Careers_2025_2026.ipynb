# Data Visualization 1: Salary Growth Across Experience Levels

## Objective

The first visualization examines how compensation changes as AI professionals gain experience.

Understanding salary progression helps illustrate the long-term financial opportunities available within AI careers and provides context for career planning and workforce development.

```python
# Average salary by experience level

experience_salary = (
    df.groupby('experience_level')['annual_salary_usd']
      .mean()
      .reindex([
          'Entry (0-2 yrs)',
          'Mid (3-5 yrs)',
          'Senior (6-9 yrs)',
          'Lead (10+ yrs)'
      ])
)

experience_salary
```

```python

from matplotlib.ticker import FuncFormatter

# Create figure
fig, ax = plt.subplots(figsize=(8, 6))

# Plot line
ax.plot(
    experience_salary.index,
    experience_salary.values,
    marker='o',
    linewidth=2.5,
    markersize=8
)

# Title
ax.set_title(
    'Average Annual Salary by Experience Level in AI Careers',
    fontsize=15,
    fontweight='bold',
    pad=15
)

# Axis Labels
ax.set_xlabel(
    'Experience Level',
    fontsize=12,
    labelpad=10
)

ax.set_ylabel(
    'Average Annual Salary (USD)',
    fontsize=12,
    labelpad=10
)

# Format Y-axis
ax.yaxis.set_major_formatter(
    FuncFormatter(lambda x, pos: f'${x:,.0f}')
)

# Remove gridlines
ax.grid(False)

# Remove extra borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Dynamic headroom for labels
y_min = experience_salary.min()
y_max = experience_salary.max()
padding = (y_max - y_min) * 0.18

ax.set_ylim(
    y_min - padding * 1.15,
    y_max + padding
)

# Dynamic data labels
for x, y in zip(experience_salary.index, experience_salary.values):
    ax.annotate(
        f'${y/1000:.0f}K',
        xy=(x, y),
        xytext=(0, 12),
        textcoords='offset points',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

plt.tight_layout()

# Save for PowerPoint
plt.savefig(
    'salary_by_experience_level.png',
    dpi=300,
    bbox_inches='tight'
)

plt.savefig(
    "salary_by_experience_level.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

# Visualization 3: AI Career Opportunity Matrix

## Objective

While previous visualizations examined salary and demand separately, this analysis combines both metrics to identify the most attractive AI career opportunities.

The visualization compares average annual salary and market demand across AI roles. Reference lines representing overall average salary and demand create four quadrants that help identify positions offering the strongest combination of compensation and market demand.

Roles appearing in the upper-right quadrant represent the most attractive opportunities because they combine above-average salaries with above-average demand.

```python


# Create role-level summary
role_matrix = (
    df.groupby('job_title')
      .agg({
          'annual_salary_usd': 'mean',
          'demand_score': 'mean'
      })
      .reset_index()
)

# Reference lines
salary_avg = role_matrix['annual_salary_usd'].mean()
demand_avg = role_matrix['demand_score'].mean()

# Key roles to label
highlight_roles = [
    'LLM Engineer',
    'Senior ML Engineer',
    'AI Agent Developer',
    'RAG Engineer',
    'AI Solutions Architect',
    'ML Engineer'
]

fig, ax = plt.subplots(figsize=(8, 6))

# Scatter points
ax.scatter(
    role_matrix['demand_score'],
    role_matrix['annual_salary_usd'],
    s=80,
    alpha=0.75
)

# Average lines
ax.axvline(
    demand_avg,
    linestyle='--',
    linewidth=1.5
)

ax.axhline(
    salary_avg,
    linestyle='--',
    linewidth=1.5
)

# Labels
for _, row in role_matrix.iterrows():

    if row['job_title'] in highlight_roles:

        ax.annotate(
            row['job_title'],
            (row['demand_score'], row['annual_salary_usd']),
            xytext=(5, 5),
            textcoords='offset points',
            fontsize=8
        )

# Quadrant title
ax.text(
    demand_avg + 2,
    salary_avg + 40000,
    'Best Opportunities',
    fontsize=10,
    fontweight='bold'
)

# Formatting
ax.set_title(
    'AI Career Opportunity Matrix: Salary vs Market Demand',
    fontsize=15,
    fontweight='bold',
    pad=15
)

ax.set_xlabel(
    'Market Demand Score',
    fontsize=12
)

ax.set_ylabel(
    'Average Annual Salary ($K)',
    fontsize=12
)

ax.yaxis.set_major_formatter(
    FuncFormatter(lambda x, pos: f'${x/1000:.0f}K')
)

ax.grid(False)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()

plt.savefig(
    'ai_career_opportunity_matrix.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

### Analysis

The upper-right quadrant represents AI careers that offer both above-average salaries and above-average demand. Roles such as LLM Engineer, Senior ML Engineer, AI Agent Developer, and RAG Engineer appear in this region, making them some of the most attractive career paths in the dataset.

Several specialized AI positions earn high salaries but show more moderate demand, while other roles experience strong demand without offering equivalent compensation. This suggests that both skill specialization and market demand contribute to earning potential.

Overall, the analysis highlights advanced machine learning, generative AI, and large language model roles as the strongest opportunities in today's AI job market.

# Visualization 4: AI Career Value Landscape

## Objective

Salary alone does not provide a complete picture of career value. This visualization examines three dimensions of AI careers simultaneously: market demand, annual compensation, and AI salary premium.

Bubble size represents the additional salary premium associated with AI-related skills, while position indicates overall demand and compensation. This allows us to identify roles that offer the strongest overall market value.

```python
# Create role-level summary

bubble_df = (
    df.groupby(['job_title', 'job_category'])
      .agg({
          'annual_salary_usd': 'mean',
          'demand_score': 'mean',
          'ai_salary_premium_pct': 'mean'
      })
      .reset_index()
)
```

```python

# ==================================================
# Create Opportunity Score
# ==================================================

bubble_df['opportunity_score'] = (
    bubble_df['annual_salary_usd']
    *
    bubble_df['demand_score']
)

# ==================================================
# Bubble Sizes
# ==================================================

bubble_df['bubble_size'] = (
    bubble_df['ai_salary_premium_pct']
    /
    bubble_df['ai_salary_premium_pct'].max()
) * 600 + 60

# ==================================================
# Median Values
# ==================================================

median_salary = (
    bubble_df['annual_salary_usd']
    .median()
)

median_demand = (
    bubble_df['demand_score']
    .median()
)

# ==================================================
# Top Opportunities
# ==================================================

top_roles = (
    bubble_df
    .nlargest(
        8,
        'opportunity_score'
    )
)

# ==================================================
# Dynamic Colors
# ==================================================

categories = sorted(
    bubble_df['job_category'].unique()
)

colors = plt.cm.tab20(
    np.linspace(
        0,
        1,
        len(categories)
    )
)

category_color_map = dict(
    zip(categories, colors)
)

# ==================================================
# Figure
# ==================================================

fig, ax = plt.subplots(
    figsize=(12, 8)
)

# ==================================================
# Plot Bubbles
# ==================================================

for category in categories:

    subset = bubble_df[
        bubble_df['job_category']
        == category
    ]

    ax.scatter(
        subset['demand_score'],
        subset['annual_salary_usd'],
        s=subset['bubble_size'],
        color=category_color_map[category],
        alpha=0.60,
        edgecolors='black',
        linewidth=0.6
    )

# ==================================================
# Dynamic Labels
# ==================================================

texts = []

for _, row in top_roles.iterrows():

    texts.append(

        ax.text(
            row['demand_score'],
            row['annual_salary_usd'],
            row['job_title'],
            fontsize=8,
            fontweight='bold'
        )

    )

adjust_text(
    texts,
    arrowprops=dict(
        arrowstyle='-',
        color='gray',
        lw=0.5
    )
)

# ==================================================
# Median Lines
# ==================================================

ax.axhline(
    median_salary,
    linestyle='--',
    color='gray',
    linewidth=1.2
)

ax.axvline(
    median_demand,
    linestyle='--',
    color='gray',
    linewidth=1.2
)

# ==================================================
# Quadrant Labels
# ==================================================

box_style = dict(
    boxstyle='round,pad=0.4',
    facecolor='white',
    edgecolor='lightgray',
    alpha=0.90
)

ax.text(
    bubble_df['demand_score'].min() + 1,
    bubble_df['annual_salary_usd'].max() - 3000,
    'High Salary\nLow Demand',
    fontsize=10,
    fontweight='bold',
    color='darkred',
    bbox=box_style
)

ax.text(
    bubble_df['demand_score'].max() - 8,
    bubble_df['annual_salary_usd'].max() - 3000,
    'High Salary\nHigh Demand',
    fontsize=10,
    fontweight='bold',
    color='darkgreen',
    bbox=box_style
)

ax.text(
    bubble_df['demand_score'].min() + 1,
    bubble_df['annual_salary_usd'].min() + 5000,
    'Low Salary\nLow Demand',
    fontsize=10,
    fontweight='bold',
    color='darkorange',
    bbox=box_style
)

ax.text(
    bubble_df['demand_score'].max() - 8,
    bubble_df['annual_salary_usd'].min() + 5000,
    'Low Salary\nHigh Demand',
    fontsize=10,
    fontweight='bold',
    color='navy',
    bbox=box_style
)


# ==================================================
# Title
# ==================================================

ax.set_title(
    'AI Career Value Landscape: Demand, Salary, and Premium (2025–2026)',
    fontsize=16,
    fontweight='bold',
    pad= 25
)

ax.text(
    0.5,
    1.02,
    'Bubble size represents average AI salary premium (%)',
    transform=ax.transAxes,
    ha='center',
    fontsize=10,
    style='italic'
)

# ==================================================
# Axis Labels
# ==================================================

ax.set_xlabel(
    'Demand Score',
    fontsize=12
)

ax.set_ylabel(
    'Average Annual Salary ($K)',
    fontsize=12
)

ax.yaxis.set_major_formatter(
    FuncFormatter(
        lambda x, pos: f'${x/1000:.0f}K'
    )
)

# ==================================================
# Remove Gridlines
# ==================================================

ax.grid(False)


# ==================================================
# Layout
# ==================================================

plt.tight_layout(
    rect=[0, 0.06, 0.88, 1]
)

# ==================================================
# Save Figure
# ==================================================

plt.savefig(
    'ai_career_value_landscape.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

### Analysis

- AI Engineering roles occupy the upper-right region of the chart, indicating the strongest combination of market demand and salary.
- Senior ML Engineer, LLM Engineer, and AI Agent Developer emerge as the most attractive career paths due to their high demand scores and compensation levels.
- AI Solutions Architect offers the highest average salary despite slightly lower demand than leading engineering roles.
- Larger bubbles are concentrated among advanced technical roles, suggesting that specialized AI skills command a significant salary premium.
- Governance, compliance, and business-focused roles generally show lower demand and lower compensation compared with technical AI positions.
- Emerging fields such as Multimodal AI Engineering, RAG Engineering, and AI Agent Development demonstrate strong market demand and above-average salaries.
- Overall, the visualization reveals a positive relationship between market demand and compensation, indicating that highly sought-after AI skills are rewarded with higher salaries.
